## Introduction

- The aim of this notebook is to predict the correct spelling of the words ,  without using  NLP library.
- It takes `big.txt` as a reference for checking the  correct spellings.

## Import Library

```python
import re
import pandas as pd 
from tqdm import tqdm
```

## Loading Dataset

```python
with open('big.txt','r') as file:
    lines = file.readlines()
    words = []
    for line in lines:
        words += re.findall('\w+',line.lower())
        
len(words)     
```

```python
words
```

```python
with open('big.txt','r') as file:
    lines = file.readlines()
    words1 = []
    for line in lines:
        words1 += line.split(' ')
        
len(words1) 
```

```python
words1[:100]
```

By using `Regular expression(re)` we are getting only solid words. But when we use `.split('')` method, we are getting escape sequences and other words

### 1.Finding Unique Words

```python
#finding unique words
vocab= set(words)
```

```python
len(vocab)
```

```python
vocab
```

### 2.Finding Probablity Distribution

```python
word_proab = {}

for word in tqdm(vocab):
    word_proab[word] = words.count(word)/len(words)
```

```python
len(word_proab)
```

```python
word_proab
```

### 3.Text Preprocessing
        a. Splitting
        b. Delete
        c. Swap
        d. Replace
        e. Insert

```python
#Splitting
def split(word):
    parts = []
    for i in range(len(word)+1):
        parts += [(word[:i],word[i:])]
    return parts
```

```python
split('wonderful')
```

```python
#delete
def delete(word):
    comb = []
    for l,r in split(word):
        comb.append(l + r[1:])
    return comb

delete('wonderfull')
```

```python
#swap
def swap(word):
    output = []
    for l,r in split(word):
        if len(r) > 1:
            output.append(l + r[1] + r[0] + r[2:])
    return output

swap('wonderful')
```

```python
# replace
def replace(word):
    characters = 'abcdefghijklmnopqrstuvwxyz'
    output = []
    for l, r in split(word):
        for char in characters:
            output.append(l + char + r[1:])
    return output
replace('the')
```

```python
# replace
def insert(word):
    characters = 'abcdefghijklmnopqrstuvwxyz'
    output = []
    for l, r in split(word):
        for char in characters:
            output.append(l + char + r)
    return output
replace('the')
```

### 4.Finding the Prediction(1-level)

```python
#combining possible words

def edit(word):
    return list(set(insert(word) + delete(word) + swap(word) + replace(word)))

```

```python
sugessted_word = edit('lovae')
sugessted_word[:50]
```

```python
def spell_check(word,count = 5):
    output = []
    sugessted_word = edit(word)
    for word in tqdm(sugessted_word):
        if word in word_proab.keys():
            output.append([word,word_proab[word]])

    return list(pd.DataFrame(output,columns=['word','prob']).sort_values(by='prob',ascending=False).head(count)['word'].values)       
```

```python
spell_check('familr')
```

```python
spell_check('ang')
```

```python
spell_check('fdmilr')
```

### 5.Finding the Prediction(2-level)

```python
def edit_2(word):
    suggested_words = edit(word)
    for e1 in tqdm(edit(word)):
        suggested_words += edit(e1) #second level edit
    return list(set(suggested_words))
```

```python
def spell_check_advance(word,count = 5):
    output = []
    sugessted_word = edit_2(word)
    for word in tqdm(sugessted_word):
        if word in word_proab.keys():
            output.append([word,word_proab[word]])

    return list(pd.DataFrame(output,columns=['word','prob']).sort_values(by='prob',ascending=False).head(count)['word'].values)       
```

```python
spell_check_advance('fdmilr')
```

```python
spell_check_advance('ang')
```
